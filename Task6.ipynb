# #  <span style='color:red'>Importing all the required Libraries </span>

```python
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn.datasets as datasets
```

# #  <span style='color:red'>Importing Iris dataset from sklearn library</span>

```python
iris=datasets.load_iris()
```

```python
# Forming the iris dataframe
df=pd.DataFrame(iris.data, columns=iris.feature_names)
df.head(10)
```

```python
y=iris.target
y
```

```python
print(iris.target_names)
```

# # <span style='color:red'>checking null value </span>

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.shape
```

# # <span style= 'color:red' >plot a pairplot to gain more insights from the data</span>

```python
sns.pairplot(df)
```

# # <span style ='color :red'> Model Creation</span >

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size = 0.2, random_state = 6)
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

# # <span style ='color :red'>Visualizing the Decision Tree</span>

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

fig = plt.figure(figsize = (10,10))
plot_tree(classifier, feature_names = iris.feature_names, class_names=iris.target_names, filled = True, rounded = True)
plt.show()
```

```python
# Predicting the model on test data 
y_pred = classifier.predict(X_test)
y_pred
```

```python
# Accuracy
from sklearn import metrics
print("Accuracy: ", metrics.accuracy_score(y_test, y_pred))
```
